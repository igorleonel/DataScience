## Tratando os Dados

No caso desse conjunto de dados especfico, o ideal  analisar os dados apenas do perodo em que se tem dados. Especialmente em algumas variveis onde o volume  muito grande, qualquer tipo de preenchimento poderia enviesar os dados de forma que a anlise deixasse de ser relevante.

Para outros casos, podemos considerar as seguintes hipteses:

* Excluir
  * Se os dados ausentes esto em pequeno nmero,ocorrem aleatoriamente, e a ausncia no carrega significado,  melhor excluir a linha. No caso da coluna, se ainda for possvel analisar alguma parte dela, use-a, como  o caso aqui. Mas para algumas situaes, o ideal  excluir a coluna.

* Preencher
  * Preencher as entradas com dados ausentes com valores estatsticos como a mdia, mediana, moda ou zeros.
  * A mdia  mais til quando a distribuio dos dados  normal. Em dados com distribuio mais enviesada (*skewed*), a mediana  uma soluo mais robusta, pois ela  menos sensvel a outliers.
  * Uma `Regresso Linear` tambm pode ser til, apesar de sensvel a outliers, podem nos ajudar a inserir valores que nos ajudem.
  * Indetificar a entrada ausente com algum valor que indique isso pode ser mais informativo, quando a ausncia representa valor. Por exemplo, em dados numricos preencher com zero, e em categricos criar uma categoria "Desconhecido". Ateno, pois os zeros no podem ser levados em considerao em anlises estatsticas.

## Limpando Outliers

Para tratar dos outliers desse conjunto de dados, iremos analisar a distribuio estatstica, plotar boxplots e calcular os limites utilizando a regra do IQR Score.

Primeiramente, vamos lembrar o que  o IQR.

O IQR  calculado subtraindo o Terceiro Quartil (75%) pelo Primeiro Quartil (25%).

# IQR = Q3 - Q1

Vamos dar uma olhada nos nossos dados e ver o que identificamos.

```
# verificando a distribuio estatstica
df_nyc.describe()
```

Aqui, algumas coisas j chamam a nossa ateno, como por exemplo:

* A varivel `price` tem o mnimo em 0.
* Lembrando que a varivel `price` trata do preo da diria dos imveis em moeda local (USD), estamos vendo que o Q3 est em 175 dlares, mas o mximo est em 10 mil drales. Claramente, h outliers por aqui.
* A varivel `minimum_nights` tem como seu mximo o valor 1250, sendo que o Q3 est em 6. Claramente temos outliers nessa varivel.
* As variveis `number_of_reviews`, `calculated_host_listings_count` e `availability_365` tambm podem conter outliers, mas no vamos nos preocupar com elas agora.

```
# verificando as distribuies
df_nyc.hist(figsize=(20, 15), grid=False)
```

Verificando os histogramas, conseguimos ver claramente que temos outliers presentes. Para trat-los vamos seguir os seguintes passos:

* Definir o Q1 e Q3 para as variveis que sero limpas.
* Calcular o IQR para as variveis.
* Definir o limite superior e inferior para cortar os outliers.
* Remover os outliers.


```
# identificando os outliers para a varivel price
q1_price = df_nyc.price.quantile(.25)
q3_price = df_nyc.price.quantile(.75)
IRQ_price = q3_price - q1_price
print('IRQ da varivel price: ', IRQ_price)

# definindo os limites                                       
sup_price = q3_price + 1.5 * IRQ_price
inf_price = q1_price - 1.5 * IRQ_price

print('Limite superior de price: ', sup_price)
print('Limite inferiorde price: ', inf_price)
```

Aqui podemos ver que, apesar de no termos outliers na parte inferior, continuamos tendo valores iguais a zero, que precisam ser tratados.

Vamos plotar um boxplot para visualizarmos a diferena feita pela limpeza.

```
# verificando o conjunto original
fig, ax = plt.subplots(figsize=(15, 3))
df_nyc.price.plot(kind='box', vert=False)
ax.set_title('Dataset Original - price')
plt.show()
print('O dataset possui {} colunas'.format(df_nyc.shape[1]))
print('{} Entradas acima de 335.5'.format(len(df_nyc[df_nyc.price > sup_price])))
print('Representam {:.2f}% do dataset'.format((len(df_nyc[df_nyc.price > sup_price]) / df_nyc.shape[0]) * 100))
```

```
df_nyc.shape
```

```
# identificando os outliers para a varivel minimum_nights
q1_minimum_nights = df_nyc.minimum_nights.quantile(.25)
q3_minimum_nights = df_nyc.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print('IQR da varivel minimum_nights: ', IQR_minimum_nights)

# definindo os limites                                       
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights
inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights

print('Limite superior de minimum_nights: ', sup_minimum_nights)
print('Limite inferior de minimum_nights: ', inf_minimum_nights)
```

```
# verificando o conjunto original
fig, ax = plt.subplots(figsize=(15,3))
df_nyc.minimum_nights.plot(kind='box', vert=False);
ax.set_title('Dataset Original - minimum_nights')
plt.show()
print("O dataset possui {} colunas".format(df_nyc.shape[0]))
print("{} Entradas acima de 335.5".format(len(df_nyc[df_nyc.minimum_nights > sup_minimum_nights])))
print("Representam {:.2f}% do dataset".format((len(df_nyc[df_nyc.minimum_nights > sup_minimum_nights]) / df_nyc.shape[0])*100))
```

```
# limpando o dataset
df_clean = df_nyc.copy()

df_clean.drop(df_clean[df_clean.price > 335.5].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.price == 0.0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 4].index, axis=0, inplace=True)

print('Shape antes da limpeza: ', df_nyc.shape)
print('Shape aps a limpeza: ', df_clean.shape)
```

```
# plotando novamente o boxplot original
fig, ax = plt.subplots(figsize=(15,3))
df_nyc.minimum_nights.plot(kind='box', vert=False);
ax.set_title('Dataset Original - minimum_nights')
plt.show()
print("O dataset possui {} colunas".format(df_nyc.shape[0]))
print("{} Entradas acima de 335.5".format(len(df_nyc[df_nyc.minimum_nights > sup_minimum_nights])))
print("Representam {:.2f}% do dataset".format((len(df_nyc[df_nyc.minimum_nights > sup_minimum_nights]) / df_nyc.shape[0])*100))
```

```
# verificando o conjunto limpo
fig, ax = plt.subplots(figsize=(15,3))
df_clean.price.plot(kind='box', vert=False);
ax.set_title('Dataset Limpo - price')
plt.show()
print("Tamanho original: {} entradas".format(df_nyc.shape[0]))
print("Tamanho atual: {} entradas".format(df_clean.shape[0]))
```

Aqui podemos ver o resultado da limpeza.

O boxplot mostra alguns dados como outliers, entretanto, esses esto sendo calculados em relao ao novo dataset, e nossa limpeza levou em considerao os quartis do dado original.

## Informaes Importantes

* Modelos lineares so mais sensveis aos outliers. Ao trabalhar com modelos desse tipo  essencial que o trabalho com outliers seja feito com bastante ateno. (Linear Regression, Logistic Regression)

* Modelos baseados em rvores de deciso so menos sensvels a outliers. (Decision Trees, Random Forest, SVM, KNN)

* Tcnicas de pr-processamento como Normalizao e Padronizao ajudam em relao aos outliers.
